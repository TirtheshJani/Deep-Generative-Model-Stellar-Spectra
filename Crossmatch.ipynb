```python
def list_files(directory, extension):
    saved = getcwd()
    chdir(directory)
    it = glob('*.' + extension)
    chdir(saved)
    return it
```

```python
targetdir_apg = "//arc/projects/k-pop/spectra/apogee/dr17"

selected_files_apg = list_files(targetdir_apg, 'fits')

selected_file_apg = int(len(selected_files_apg))
print("Number of files: " + str(selected_file_apg))
```
